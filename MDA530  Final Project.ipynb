# Bar Chart

```python
vg_bar= vg[['Genre', 'Global_Sales']]
vg_ed =vg_bar.groupby(['Genre'])['Global_Sales'].sum().reset_index()
plt.bar(vg_ed.Genre, vg_ed.Global_Sales)
plt.title('Genre vs Global Sales')
plt.xlabel('Genre')
plt.ylabel('Global_Sales')
plt.xticks(rotation=90)

plt.show()
```

```python
fig, ax = plt.subplots()
g=ax.barh(vg_ed['Genre'], vg_ed['Global_Sales'], color='Green')
ax.set_xlabel("Global Sales")
ax.set_ylabel("Genre")
ax.bar_label(g, label_type="center")
```
